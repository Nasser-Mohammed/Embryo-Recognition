```python
def create_lenet():
    model = nn.Sequential(
        
    nn.Conv2d(1, 6, 5, padding = 2),
    nn.ReLU(),
    nn.AvgPool2d(2, stride = 2),
    
    nn.Conv2d(6, 16, 5, padding = 0),
    nn.ReLU(),
    nn.AvgPool2d(2, stride = 2), 
    
    nn.Flatten(),
    nn.Linear(76032, 120),
    nn.ReLU(),
    nn.Linear(120, 84),
    nn.ReLU(),
    nn.Linear(84, 7)
)
    return model
```

```python
def validate(model, data):
    total = 0
    correct = 0
    for i, (images, labels) in enumerate(data):
        x = model(images)
        value, pred = torch.max(x, 1)
        total += x.size(0)
        #useless, labels_index = torch.max(labels,1)
        correct += torch.sum(pred == labels)
    
    return correct*100./total
```

```python
def train(lr, epochs, device = "cpu"):
    accuracies = []
    cnn = create_lenet().to(device)
    cec = nn.CrossEntropyLoss()
    optimizer = optim.Adam(cnn.parameters(), lr = lr)
    max_accuracy = 0
    
    for epoch in range(epochs):
        ttl = 0
        crct = 0
        trainAccuracy = 0.0
        
        for i, (images, labels) in enumerate(train_dl):
            optimizer.zero_grad()
            pred = cnn(images)
            ttl += pred.size(0)
            loss = cec(pred, labels)
            loss.backward()
            optimizer.step()
            crct += (pred.argmax(1) == labels).sum()
        print(f"Train accuracy for epoch: {epoch} is {crct*100./ttl}")
        accuracy = float(validate(cnn, val_dl))
        accuracies.append(accuracy)
        if accuracy > max_accuracy:
            best_model = copy.deepcopy(cnn)
            max_accuracy = accuracy 
            print("Saving Best Model With Accuracy: ", accuracy)
        print("Epoch: ", epoch+1, "Accuracy: ", accuracy, "%")
    #plt.plot(accuracies)
    return best_model, accuracies, lr
```

```python
lenet = train(0.001, 15)
```

```python
def predict_dl(model, data):
    y_pred = []
    y_true = []
    for i, (images, labels) in enumerate(data):
        x = model(images)
        value, pred = torch.max(x, 1)
        pred = pred.data.cpu()
        y_pred.extend(list(pred.numpy()))
        y_true.extend(list(labels.numpy()))
    return np.array(y_pred), np.array(y_true)
```

```python
y_pred, y_true = predict_dl(lenet[0], test_dl)
```
